```python
#classic

#import libraries
from sklearn.datasets import make_regression
import matplotlib.pyplot as plt

# data
data = make_regression(n_samples=40, n_features=1, noise=10, random_state=42)
x_data, y_data = data[0], data[1]

# model
def classic_program(x): return x*70

y_pred = [classic_program(x) for x in x_data]

# plotting
plt.scatter(x_data, y_data, label='Real datasets')
plt.plot(x_data, y_pred, 'r', label='Program')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.title('x*70')
```

```python
#adding "weight"

from sklearn.datasets import make_regression
import matplotlib.pyplot as plt

# data
data = make_regression(n_samples=40, n_features=1, noise=10, random_state=42)
x_data, y_data = data[0], data[1]



def weight_assigned_program(x, w): return w*x*70

w = 1.3

y_pred = [weight_assigned_program(x, w) for x in x_data]

# plotting
plt.scatter(x_data, y_data, label='Real datasets')
plt.plot(x_data, y_pred, 'r', label='Program with weight')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.title(f'{w}*x*70')
```

```python
#machine learning

from sklearn.datasets import make_regression
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt

# data
data = make_regression(n_samples=40, n_features=1, noise=10, random_state=42)
x_data, y_data = data[0], data[1]

# regression
regression = LinearRegression().fit(x_data, y_data)
y_pred = regression.predict(x_data)

# plotting
plt.scatter(x_data, y_data, label='Real datasets')
plt.plot(x_data, y_pred, 'r', label='ML program')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
equation = f'{regression.coef_[0]:.3f}x+{regression.intercept_:.3f}'
plt.title(equation)
```

```python
#deep learning

from sklearn.datasets import make_regression
from sklearn.neural_network import MLPRegressor
import matplotlib.pyplot as plt

# data
data = make_regression(n_samples=40, n_features=1, noise=10, random_state=42)
x_data, y_data = data[0], data[1]

# regression
regression = MLPRegressor(max_iter=2000).fit(x_data, y_data)
y_pred = regression.predict(x_data)


# plotting
plt.figure(figsize=(10, 8))
plt.scatter(x_data, y_data, label='Haqiqiy data sets')
plt.plot(x_data, y_pred, 'r', label='DL dastur')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
```
